$$SS_{xx} = \sum(x_i - \bar{x})^2$$

$$RSE = \sqrt{\frac{RSS}{n-2}}$$

$$SE(\hat{\beta}_1) = \frac{RSE}{SS_{xx}}$$

$$SE(\hat{\beta}_0) = RSE \sqrt{\frac{1}{n} + \frac{\bar{x}^2}{SS_{xx}}}$$

```python
# Equation for Sum of squared deviations for x
SSxx = SSxx = np.sum((df['dna_richness'] - df['dna_richness'].mean())**2)

# Equation for Residual sum of squares to use in RSE
RSS = np.sum((df['trad_richness'] - (beta_0 + beta_1 * df['dna_richness']))**2)

# Residual Squared Error is equal to sqrt(RSS / (number of rows - 2))
RSE = (RSS / (len(df) - 2))**0.5

# Reference above equations to calculate SE for beta1 and beta0
SE_beta_1 = (RSE / ((SSxx)**.5))
SE_beta_0 = (RSE * ((1/len(df)) + (df['dna_richness'].mean()**2 / SSxx))**0.5)

print(f"RSE      = {RSE:.4f}  (species)")
print(f"SE(\u03b2\u2081)   = {SE_beta_1:.4f}")
print(f"SE(\u03b2\u2080)   = {SE_beta_0:.4f}")
```
